## Task 4

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, silhouette_score, matthews_corrcoef, mean_squared_error
import xgboost as xgb
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import KFold, cross_val_score
```

```python
label = pd.DataFrame()
label["name"] = data.index
label["PAM50"] = 1
```

```python
for i in range(len(clinical)):
    if clinical['pam50'][i]=="Basal":
        label["PAM50"][i] = 0
```

```python
label
```

```python
x = data
y = list(label["PAM50"])
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 123)
```

```python
rf = RandomForestClassifier(random_state=123)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)

print("Accuracy  : ", accuracy_score(y_test, rf_pred))
print("Recall    : ", recall_score(y_test, rf_pred))
print('\nMCC: %f' % (matthews_corrcoef(y_test, rf_pred)))
confusion_matrix(y_test, rf_pred)
```

## Task 7

### data preprocessing

```python
label2 = pd.DataFrame()
label2["name"] = data.index
label2["live_years"] = 1
```

```python
remove_alive = []
for i in range(len(clinical)):
  if clinical['vital_status'][i]=="Alive":
    remove_alive.append(i)
remove_alive
```

```python
label2.drop(index=remove_alive,inplace=True)
```

```python
label2
```

```python
for i in range(len(clinical)):
    if clinical['vital_status'][i]=="Dead":
        label2["live_years"][i] = clinical['year_of_death'][i] - clinical['year_of_birth'][i]
```

```python
label2.reset_index(drop=True, inplace=True)
```

```python
label2
```

```python
remove_nan = []
for i in range(len(label2)):
  if pd.isna(label2['live_years'][i]):
    remove_nan.append(i)
```

```python
label2.drop(index=remove_nan, inplace=True)
```

```python
label2.reset_index(drop=True, inplace=True)
```

```python
label2
```

```python
label2_name = label2['name'].unique()
```

```python
data_new = data.reset_index()
```

```python
data_new
```

```python
data_remove = []
for i in range(len(data_new)):
  if data_new['index'][i] not in label2_name:
    data_remove.append(i)
```

```python
data_new.drop(index=data_remove, inplace=True)
```

```python
data_new
```

```python
data_new.set_index('index' ,inplace=True)
```

```python
data_new
```

```python
x = data_new
y = list(label2["live_years"])
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 123)
```

```python
xg = xgb.XGBRegressor(random_state=123)
xg.fit(X_train, y_train)  
y_pre = xg.predict(X_test)

mse = mean_squared_error(y_test, y_pre)

print("MSE: ", mse)
```

```python
importance = pd.DataFrame()
importance["index"] = list(data_new.columns)
importance["importance"] = list(xg.feature_importances_)
importance.sort_values(by="importance", ascending=False).head()
```

```python
gene_index = []
ls = importance.sort_values(by="importance", ascending=False)["index"][:20].to_list()
for i in ls:
    gene_index.append(i)
```

```python
key_genes = []
for i in range(len(gene_index)):
  for j in range(len(df)):
    if gene_index[i] == df.index[j]:
      key_genes.append(df['Unnamed: 0'][j])
```
